```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out outliers and return the modified dataframe
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply the function to the 'Rent' column
df = remove_outliers(df, 'Rent')

# Apply the function to the 'Area_in_sqft' column
df = remove_outliers(df, 'Area_in_sqft')

# Apply the function to the 'Rent_per_sqft' column
df = remove_outliers(df, 'Rent_per_sqft')

# Display the shape of the data to see how many rows are left after removing outliers
print("Data shape after removing outliers:", df.shape)
```
